Crear la clase Transporte, con 3 atributos (tipo, capacidad, costo), 2 funciones leerdatos(), mostrardatos() y 2 objetos (transporte1, transporte2).

```python
# Act25 Clase Transporte Marco 1440
# Alumno: Marco Morquecho | ID: 1440 | NL: 39 | Ejercicio: 42
# Descripción: Implementación interactiva de POO con la Clase Transporte.

class Transporte:
    # Atributos: tipo, capacidad, costo

    def leerdatos(self):
        """Captura de datos interactiva desde el teclado"""
        print(f"\n{'='*10} REGISTRO DE TRANSPORTE {'='*10}")
        self.tipo = input("Ingrese el tipo de transporte (ej. Camión, Avión): ")
        self.capacidad = input("Ingrese la capacidad (ej. 20 Ton, 150 Pasajeros): ")
        self.costo = input("Ingrese el costo operativo: ")

    def mostrardatos(self):
        """Despliegue de reporte técnico de la instancia"""
        print("\n" + "·"*40)
        print(f"FICHA TÉCNICA DE UNIDAD")
        print(f"UNIDAD:     {self.tipo}")
        print(f"CAPACIDAD:  {self.capacidad}")
        print(f"COSTO:      ${self.costo}")
        print("·"*40)

# --- PROGRAMA PRINCIPAL ---
print("SISTEMA DE GESTIÓN LOGÍSTICA V3")

# Creación de 2 objetos (instancias)
transporte1 = Transporte()
transporte1.leerdatos()

transporte2 = Transporte()
transporte2.leerdatos()

# Salida de resultados
print("\n--- RESUMEN DE FLOTA REGISTRADA ---")
transporte1.mostrardatos()
transporte2.mostrardatos()

print(f"\nRealizado por: Marco Morquecho | NL: 39+3= 42 | ID: 1440")
```
